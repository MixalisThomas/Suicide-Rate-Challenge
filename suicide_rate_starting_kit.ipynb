# Introduction <a name="introduction"></a>

According the World Health Organization, about 800 000 people die due to suicide every year, which is one person every 40 seconds.

Suicide is a global phenomenon. In fact, 79% of suicides occurred in low and middle-income countries in 2016. Suicide accounted for 1.4% of all deaths worldwide, making it the 18th leading cause of death in 2016.

In fact, more people are killed by suicide than any other type of violence, including armed conflict,each year. In Europe, the number of suicide is higher than the number of deaths due to road accident.


# Motivation

Every suicide is a tragedy that affects families, communities and entire countries and has long-lasting effects on the people left behind. Suicide occurs throughout the lifespan and was the second leading cause of death among 15-29 year-olds globally in 2016.

Suicide does not just occur in high-income countries, but is a global phenomenon in all regions of the world. In fact, over 79% of global suicides occurred in low- and middle-income countries in 2016.

Suicide is a serious public health problem. However, suicides are preventable with timely, evidence-based and often low-cost interventions. For national responses to be effective, a comprehensive suicide prevention strategy is needed. 

Nevertheless, this response has to begin with a precise analysis of the suicide factors of a country. Thus, the aim of this challenge is to identify and analyze the different components that may affect the suicide rate. Lastly, we want to predict the suicide rate per country of the year 2016. 


# Definition of the problem 

+ Aim : predict the suicide rate per country for 2016 
+ Scoring : For our metric we use the Relative Root Mean Squared Error (RRMSE):<br>
$$RRMSE = \sqrt{\frac{1}{n}\sum\limits_{i=1}^{n}\frac{(Y_i - \hat{Y_i})^2}{Y_i}} $$
$Y_i$ = observed values, <br>
$\hat{Y_i}$ = predicted values, <br>
$n$ = number of data points

# The features
Below you will find the description of each feature available for our challenge: 

+ country
    - Name of the country 
+ year
    - Year of the data row
+ aged_15plus_employment_rate_percent
    - Percentage of employment rate of the people over 15
+ at_least_basic_sanitation_overall_access_percent
    - People using safely managed sanitation services (% of population)
+ bwelfr_idea
    - IDEA Basic Welfare index
    - The provision of basic welfare is measured using standard observable human development indicators: infant mortality rate, life expectancy, kilocalories per person per day, literacy rate and average years of schooling. In addition, two expert-based indicators from V-Dem were included that assess whether everyone in a given society has access to basic education and health care. All the indicators reflect the extent to which the basic needs of the population are being met. The seven indicators ( infant mortality rate, life expectancy, kilocalories per person per day, literacy, mean years of schooling, educational equality  and health equality) were aggregated into the basic welfare sub- component using IRT.
+ cell_phones_total
    - Total number of cellphone subscription in the country in a given year 
+ child_mortality_0_5_year_olds_dying_per_1000_born: 
    - Death of children under 5 years old per 1000 live births
+ children_per_woman_total_fertility
    - Average number of children per woman
+ demox_eiu	
    - "This democracy index is using the data from the Economist Inteligence Unit to express the quality of democracies as a number between 0 and 100. It's based on 60 different aspects of societies that are relevant to democracy universal suffrage for all adults, voter participation, perception of human rights protection and freedom to form organizations and parties. The democracy index is calculated from the 60 indicators, divided into five ""sub indexes"", which are:
    1. Electoral pluralism index;
    2. Government index;
    3. Political participation indexm;
    4. Political culture index;
    5. Civil liberty index.
    The sub-indexes are based on the sum of scores on roughly 12 indicators per sub-index, converted into a score between 0 and 100. (The Economist publishes the index with a scale from 0 to 10, but Gapminder has converted it to 0 to 100 to make it easier to communicate as a percentage.)"
+ freligion_idea 
    - IDEA Freedom of religion index
    - A separate sub-component index comprised of two V-Dem general indicators on religious freedom based on expert surveys and a similarly broad in-house coded variable from CLD. The three indicators that were aggregated into the freedom of religion sub-component using BFA are i. Freedom of religion, ii.Religious organization repression and iii)freedom of thought, conscience and religion
+ gini
    - Gini coefficient: index to measure the degree of income inequality in a population. The higher the index, more inequality is present. 
+ hdi_human_development_index	
    - Human Development Index (HDI) 
+ income_per_person_gdppercapita_ppp_inflation_adjusted	
    - GDP/Capita adjusted for different purchansing powers (in 2011 international dolars, PPP based on 2011) 
+ inflation_annual_percent	
    - Inflation as measured by the annual growth rate of the GDP implicit deflator shows the rate of price change in the economy as a whole. The GDP implicit deflator is the ratio of GDP in current local currency to GDP in constant local currency.
+ internet_users	
    - Internet users are individuals who have used the Internet (from any location) in the last 3 months. The Internet can be used via a computer, mobile phone, personal digital assistant, games machine, digital TV etc.

    - Column to be used in the label generation in our challenge (suicide rate).

This data comes from gampminder and [The Institute for Health Metrics and Evaluation (IHME)](http://www.healthdata.org/about)The Institute for Health Metrics and Evaluation (IHME) and has some extreme missing datas filled with estimates where data is impossible to find.

It is calculated without making deductions for depreciation of fabricated assets or for depletion and degradation of natural resources. Data are in constant 2010 U.S. dollars. Dollar figures for GDP are converted from domestic currencies using 2010 official exchange rates.
+ urban_population_percent_of_total	
    - What % of the population is considered to be living in an urban area 
+ GHED_CHEGDP_SHA2011
    - Current health expenditure as a share of GDP provides an indication on the level of resources channelled to health relative to other uses. It shows the importance of the health sector in the whole economy and indicates the societal priority which health is given measured in monetary terms.

# Workflow

The work flow is composed of two essential elements that make up the submission: the feature engineering part and the regressor itself. The former is really important as it is responsible for getting the best variable to better the model. The latter is the learning model itself; thus, indispensable to predict and it should predict the suicide rate as accurately as possible. So, in order to have a good learning model, one should extract and create some features and feed them to the predictor.

```python
cat_col = ['country']
    
categorical_transformer = Pipeline(steps=[
        ('encode', OrdinalEncoder())
    ])
    

preprocessor = ColumnTransformer(
        transformers=[
            #('num', numeric_transformer, num_cols),
            ('cat', categorical_transformer, cat_col),
        ], remainder='passthrough')
    

regressor = RandomForestRegressor(
            n_estimators=5, max_depth=50, max_features="auto", n_jobs=1)
    

pipeline = Pipeline(steps=[
        ('preprocessing', preprocessor),
        ('regressor', regressor)
    ])

def rrmse(y_pred,y):
    return np.sqrt(1/len(y_pred)*np.sum((1-y_pred/y)**2))


def get_score(X_train, Y_train, X_test, Y_test):
    rf = pipeline.fit(X_train, Y_train)
    y_pred = rf.predict(X_test)
    print("The score for this estimator is ",rrmse(y_pred,Y_test))


get_score(X_train, Y_train, X_test, Y_test)
```

Some possible feature engineering methods are making extensive variables intensive, for instance, getting the number of cell phones per person and trying to remove any possible trend by taking first order or second order differences. 

```python
def feature_eng(X):
    
    transformed = X
    transformed['internet_users'] = transformed['internet_users']/transformed['population_total']
    transformed['cell_phones_total'] = transformed['cell_phones_total']/transformed['population_total']
    
    return transformed  

X_train2 = feature_eng(X_train) 
X_test2 = feature_eng(X_test)

get_score(X_train2, Y_train, X_test2, Y_test)
```

To make a RAMP submission you will need to create a new directory inside `submissions` folder. Name it however you like and within it, put your file named `estimator.py`, e.g. `submissions/subfolder/estimator.py`

In `estimator.py`, define a function named `get_estimator` that returns a scikit-learn pipeline or estimator that performs the desired feature extraction and regression.

Below we have an example of an estimator based on Random Forest.

```python
import os
import pandas as pd
import numpy as np

from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler, \
    OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor

N_JOBS = 1

def get_estimator():
    cat_col = ['country']
    
    categorical_transformer = Pipeline(steps=[
        ('encode', OrdinalEncoder())
    ])
    

    preprocessor = ColumnTransformer(
        transformers=[
            #('num', numeric_transformer, num_cols),
            ('cat', categorical_transformer, cat_col),
        ], remainder='passthrough')
    

    regressor = RandomForestRegressor(
            n_estimators=5, max_depth=50, max_features="auto", n_jobs=N_JOBS)
    

    pipeline = Pipeline(steps=[
        ('preprocessing', preprocessor),
        ('regressor', regressor)
    ])

    return pipeline
    
```
